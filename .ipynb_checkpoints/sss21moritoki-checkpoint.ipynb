# 一様分布のモデル

```python
def simulate_Unifrom(n = 100, a = 0, b = 100, fig_mode = False):
    # 一様分布に従う確率変数を生成
    X = np.random.uniform(a, b, n)
    
    data_array = np.array([X]).T
    df = data_array
    #df = pd.DataFrame(data_array, columns = ['process'])
    
    if fig_mode:
        fig, ax = plt.subplots()

        # plot the process X and volatility sigma
        ax.plot(X, label = 'process')
        ax.set_ylabel('process X')

        # 以下はそんなに関係ないから気にしなくていい．
        plt.gca().spines['right'].set_visible(False)
        plt.gca().spines['top'].set_visible(False)
        plt.gca().yaxis.set_ticks_position('left')
        plt.gca().xaxis.set_ticks_position('bottom')

        plt.legend()
    
    return df
```

## ニューラルネットワーク層を作る
まずは1層か2層でOK

#モデルの作成

```python
activation_1 = torch.nn.Sigmoid()
activation_2 = torch.nn.Sigmoid()
Hidden_layer = 4
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear( 1, Hidden_layer )
        self.layer2 = nn.Linear( Hidden_layer, 1 )
        

    def forward(self, x):
        # フォワードパスを定義
        x = activation_1(self.layer1(x))
        x = activation_2(self.layer2(x))
        return x
```

```python
# モデル（NeuralNetworkクラス）のインスタンス化
model = NeuralNetwork()
```

```python
y = model(DATAs_input) 
```

## 誤差関数を定義

最小化したいのは、
$ \int (|\Phi(x)-u(x)|^2) + u'(x))dx$

今は一様分布で、上の期待値をモンテカルロで数値計算して考える

```python
class CustomLoss(nn.Module):
    def __init__(self):
        super().__init__()
    
    def diff(self, x):
        h = 1e-4
        x_minus = x - h
        x_plus = x + h
        grad = ( model(x_plus) - model(x_minus) ) / (2 * h)
        return grad

    def forward(self, outputs, targets, grad):
        loss = ( ( targets - outputs ) ** 2 + grad ** 2 ).mean( )
        #loss = ( ( targets - outputs ) ** 2 + inputs.grad ** 2 ).mean
        #inputs = symbols("inputs")
        #loss = integrate(integ, (inputs, a, b))[0]
        return loss
```

```python
criterion = CustomLoss()
inputs = DATAs_input
outputs = model(inputs)
targets = DATAs_teacher
```

#最適化

```python
optimizer = optim.SGD(model.parameters(), lr = 0.1)
```

#学習

```python
epochs = 1000
```

```python
model.train()
for epoch in range(epochs):
  optimizer.zero_grad()
  outputs = model(inputs)
  loss = criterion(outputs, targets, inputs)
  loss_sum = torch.sum(loss)
  loss_sum.backward()
  optimizer.step()
```
